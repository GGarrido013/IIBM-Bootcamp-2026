# 10.1 The Derivative
The derivative of a function can be geometrically interpreted as the slope of the curve of a function $f(x)$. The slop is the rate of change at a given point. For example, acceleration is the derivative of speed, indicating a rate of change in speed at a given point.

It can be defined as

$$
 \dfrac{df(x)}{dx} = f'(x) = \lim_{h \to 0} \dfrac{f(x + h) - f(x)}{h}
$$

Where the derivative of a function at a point $c$ is:

$$
 f'(c) = \lim_{x \to c} \dfrac{f(x) - f(c)}{x -c}
$$

![](https://raw.githubusercontent.com/cgvalle/IIBM-BootCamp-2024/main/assets/D4/slope_plot.png)

## **What?!**

For now, just remember that the derivative is the slope of a function and let's try to look for a simpler example like a linear function:

$$
f(x) = 2x + 4
$$

The slop of that linear function is:


$$
m = \dfrac{f(x_2) - f(x_1)}{x_2 - x_1}
$$


Using python, we can find the slop by applying the previous formula


````python
# Definition of the function
def f(x):
    return 2*x + 4

# Slope of the f(x)
def slope(x1, x2):
    return (f(x2) - f(x1))/(x2 - x1)

print("Slope of f(x) for x1=1, x2=2 is       :" , slope(1, 2))
print("Slope of f(x) for x1=4, x2=2 is       :" , slope(1, 2))
print("Slope of f(x) for x1=2342, x2=2 is    :" , slope(2342, 2))
print("Slope of f(x) for x1=9, x2=2 is       :" , slope(9, 2))
print("Slope of f(x) for x1=0, x2=-23 is     :" , slope(0, -23))
````

It seems that it doesn't matter the value of $x_1$ or $x_2$. The slope is always 2. If we derivate $f(x)$:

$$
f'(x) = \dfrac{d(2x + 4)}{dx} = 2
$$

We can see that $f'(x)$ is a constant function so the value is always 2. So the result is correct, but unfortunately, this only works for linear functions since the slope is always the same value, but for a function $f(x) = cos(x/10) + (\dfrac{x}{50})^3$ will not work since it has a lot of curves.  

## but, what if we assume that $x^3$ is a linear function by zooming on it ?

Let's start by plotting it

````python
def f_x3(x):
    return np.cos(x/10) + (x/50)**3

x = np.linspace(start=-100, stop=100, num=1000)
````

````python
# Plotting the full figure
fig = plt.figure(figsize=(10, 10))  # Creating figure
ax = plt.subplot(1, 1, 1)  # Creating subplot
ax.plot(x, f_x3(x) )  # Actually plotting
plt.title("No zoom")
plt.show()
````

````python
# Zooming by changing xlim and y lim
fig = plt.figure(figsize=(10, 10))  # Creating figure
ax = plt.subplot(1, 1, 1)  # Creating subplot
ax.plot(x, f_x3(x) )  # Actually plotting
ax.set_xlim(-50, 50)
ax.set_ylim(-3, 3)
plt.title("Zoom x2")
plt.show()
````

````python
# Zooming by changing xlim and y lim
fig = plt.figure(figsize=(10, 10))  # Creating figure
ax = plt.subplot(1, 1, 1)  # Creating subplot
ax.plot(x, f_x3(x) )  # Actually plotting
ax.set_xlim(-10, 0)
ax.set_ylim(0, 1)
plt.title("Zoom x10")
plt.show()
````

````python
# Zooming by changing xlim and y lim
fig = plt.figure(figsize=(10, 10))  # Creating figure
ax = plt.subplot(1, 1, 1)  # Creating subplot
ax.plot(x, f_x3(x) )  # Actually plotting
ax.set_xlim(-8, -7)
ax.set_ylim(0.6, 0.8)
plt.title("Zoom x10000")
plt.show()
````

**We have a pseudo linear function !**, now it is possible to use $\dfrac{f(x_2) - f(x_1)}{x_2 - x_1}$ to get the slope, we only need to zoom a lot. Since zooming is not practical, we could get the same effect by using $x_1$ and $x_2$ that are separated by a small distance $h$. **The smaller $h$ the more accurate the slope is.**

#### 1. Obtain the derivative of $f(x)$ (you can do it by hand)


````python
def f_derivada(x):
    return - np.sin(x / 10)/10 + 3*(x**2)/50**3

def f(x):
    return np.cos(x / 10) + (x / 50)**3
````

#### 2. Plot $f'(x)$ in the x range {-100, 100}


````python
import numpy as np
from matplotlib import pyplot as plt

# Defining x range
x = np.linspace(-100, 100, 100)

# Creating figure
fig = plt.figure()
plt.plot(x, f_derivada(x))
plt.show()
````

#### 3. Calculate the derivate using:
$$
f'(x) = \lim_{h \to 0} \dfrac{f(x + h) - f(x)}{h}
$$
    **HINT:** $h$ is the distance between two points $x_1$ and $x_2$!. You must create a function where $h$ is a parameter that can be changed.

````python
def limit_derivative(h):
    return (f(x+h) - f(x))/h

# Defined function !
limit_derivative(2)
````

#### 4. Plot for differents values of $h$


````python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(1,1,1)
h=0.01 ;ax.plot(x, limit_derivative(h), label=f'h = {h}')
h=0.1 ;ax.plot(x, limit_derivative(h), label=f'h = {h}')
h=1 ;ax.plot(x, limit_derivative(h), label=f'h = {h}')
h=10 ;ax.plot(x, limit_derivative(h), label=f'h = {h}')
h=100 ;ax.plot(x, limit_derivative(h), label=f'h = {h}')
ax.plot(x, f_derivada(x), label='original')

plt.legend()
plt.show()
````

#### 5. Calculate the error between $f'(x)$  and the aproximated version (described at point 3. ) for different values of $h$
Error is defined as the sum of the point-wise different between the two vectors



````python
def error(h):
    # Error for different values of h
    return np.sum(np.abs((f_derivada(x) - limit_derivative(h))))

h = 0.001;  print(f"Error for h={h} is {error(h)}")
h = 0.01;  print(f"Error for h={h} is {error(h)}")
h = 0.1;  print(f"Error for h={h} is {error(h)}")
h = 1;  print(f"Error for h={h} is {error(h)}")
h = 10;  print(f"Error for h={h} is {error(h)}")
h = 100;  print(f"Error for h={h} is {error(h)}")
````

#### 6. Create a new plot of where the positives slops are green and negatives slopes are purple.


````python
dev = f_derivada(x)

positive = f(x)
positive[dev < 0] = None

negative = f(x)
negative[dev >= 0] = None

fig = plt.figure()
plt.subplot(1,1,1)
plt.plot(x, positive, color='green')
plt.plot(x, negative, color='purple')

plt.show()
````
